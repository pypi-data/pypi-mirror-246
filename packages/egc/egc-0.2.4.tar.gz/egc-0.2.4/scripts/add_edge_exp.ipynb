```python
import pandas as pd
pd.options.display.max_rows = 200
pd.options.display.max_columns = 200
```

```python
res = pd.read_csv('../results/age_add_edge.csv')
```

```python
t_res=res[['dataset','nodes_rate','add_edge_rate','ARI','NMI','ACC','Micro F1','Macro F1','Elapsed Time']]
```

```python
tmp=t_res.groupby(['dataset','nodes_rate','add_edge_rate'],as_index=False).mean().round(4)
```

```python
df_res = pd.DataFrame()
```

```python
df_res = pd.concat([df_res,tmp[tmp['dataset']=='Cora']],axis=0)
tmp[tmp['dataset']=='Cora']
```

```python
df_res = pd.concat([df_res,tmp[tmp['dataset']=='Citeseer']],axis=0)
tmp[tmp['dataset']=='Citeseer']#.sort_values(by=['ACC'],ascending=False)
```

```python
df_res = pd.concat([df_res,tmp[tmp['dataset']=='Pubmed']],axis=0)
tmp[tmp['dataset']=='Pubmed']#.sort_values(by=['ACC'],ascending=False)
```

```python
df_res = pd.concat([df_res,tmp[tmp['dataset']=='ACM']],axis=0)
tmp[tmp['dataset']=='ACM']#.sort_values(by=['ACC'],ascending=False)
```

```python
df_res = pd.concat([df_res,tmp[tmp['dataset']=='BlogCatalog']],axis=0)
tmp[tmp['dataset']=='BlogCatalog']#.sort_values(by=['ACC'],ascending=False)
```

```python
df_res = pd.concat([df_res,tmp[tmp['dataset']=='Flickr']],axis=0)
tmp[tmp['dataset']=='Flickr']#.sort_values(by=['ACC'],ascending=False)
```

```python
df_res = pd.concat([df_res,tmp[tmp['dataset']=='CoraFull']],axis=0)
tmp[tmp['dataset']=='CoraFull']#.sort_values(by=['ACC'],ascending=False)
```

```python
df_res.to_csv('../results/age_add_edge_sort.csv')
```
