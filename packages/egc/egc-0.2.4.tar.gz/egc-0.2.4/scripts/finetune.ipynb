```python
import pandas as pd
import copy
```

```python
res = pd.read_csv('../results/AGCN_0.csv')
```

```python
res
```

```python
res['model'].value_counts()
```

```python
need_cols = ['dataset','model','ARI','NMI','ACC','Micro F1','Macro F1','Elapsed Time']
t_res=res[need_cols]
```

```python
t_res = t_res[t_res['model']=='AGCN']
```

```python

t_mean = t_res.groupby(['dataset'],as_index=False).mean().round(4).T
t_mean
```

```python
t_std = t_res.groupby(['dataset'],as_index=False).std().round(4).T
t_std
```

```python
t_mean_std = copy.deepcopy(t_mean)
```

```python
for col in t_mean_std.columns.values:
    t_mean_std.iloc[1:-1,col]=t_mean_std.iloc[1:-1,col].astype(str) + '±' + t_std.iloc[1:-1,col].astype(str)
```
